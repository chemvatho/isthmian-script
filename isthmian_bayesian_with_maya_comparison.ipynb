## 1. The Research Problem

### What is the Isthmian Script?

The **Isthmian script** (also called *Epi-Olmec script*) is an undeciphered writing system from ancient Mesoamerica, used approximately 500 BCE – 500 CE in the Gulf Coast region of Mexico. The major inscriptions include:

| Inscription | Date | Location | Signs |
|-------------|------|----------|-------|
| La Mojarra Stela 1 | 156 CE | Veracruz | ~535 |
| Tuxtla Statuette | 162 CE | San Andrés Tuxtla | ~75 |
| Chiapa de Corzo Stela | 36 BCE | Chiapas | ~12 |
| Tres Zapotes Stela C | 32 BCE | Veracruz | Calendar only |

### The Decipherment Controversy

- **Justeson & Kaufman (1993)** proposed that the script encodes a **Mixe-Zoquean** language
- **Houston & Coe (2003)** showed their decipherment fails on new texts
- **Vonk (2020)** demonstrated that a **Proto-Huastecan** (Mayan) reading works equally well

### The Fundamental Problem: Corpus Too Small

Vonk (2020) formalized the **qW condition** for unique decipherment:

$$q_W = \\frac{\\text{distinct signs}}{\\text{total attestations}} \\ll 0.1$$

For Isthmian: $q_W ≈ 160/700 ≈ 0.23$ — **far exceeds the threshold**.

This means: *Traditional sign-value decipherment cannot uniquely determine the language family.*

## 2. Why Use a Bayesian Approach?

### The Key Insight

If we cannot assign phonetic values to individual signs, we can still analyze **statistical patterns** in the script that reflect properties of the underlying language:

- How long are typical word-units?
- Which signs appear at the beginning vs. end of units?
- How predictable are sign sequences?

Different language families have different **phonotactic** (sound pattern) constraints that should manifest in these statistics.

### What is Bayesian Model Comparison?

Instead of asking *"What does this sign mean?"* we ask:

> *"Given the statistical patterns in the Isthmian script, which language family's phonotactic profile best explains what we observe?"*

**Bayesian inference** provides a principled framework for this by:

1. **Prior**: What do we expect based on linguistic reconstruction?
2. **Likelihood**: How well does the observed data match each expectation?
3. **Evidence**: Combined probability of the data under each hypothesis
4. **Bayes Factor**: Ratio of evidence between competing hypotheses

### How This Contributes to the Project

| Traditional Approach | Bayesian Approach |
|---------------------|-------------------|
| Assign sign values | No sign values needed |
| Requires bilingual text | Works with script-internal data |
| Binary: deciphered or not | Graded: probability ranking |
| Subjective judgment | Explicit, reproducible assumptions |

This approach **complements** (not replaces) traditional historical-comparative reconstruction.

## 3. Glossary of Key Terms

### Isthmian Script Terms

| Term | Definition |
|------|------------|
| **MS number** | Macri-Stark catalog number for each sign (e.g., MS20, MS38) |
| **MS20** | The most frequent sign (~52 occurrences), identified by Macri (2017) as a **boundary marker** or punctuation sign that likely indicates word/phrase endings |
| **Segment** | A sequence of signs delimited by MS20 (analogous to a word or phrase) |
| **Token** | Each individual occurrence of a sign |
| **Type** | Each distinct sign (regardless of how many times it appears) |

### Linguistic Terms

| Term | Definition | Example |
|------|------------|--------|
| **Phonotactics** | Rules governing which sounds can combine in a language | English allows "str-" but not "tsr-" |
| **Syllable structure** | Pattern of consonants (C) and vowels (V) | CV = "ba", CVC = "bat", CCVC = "stop" |
| **Word-final constraints** | Which sounds can end a word | Japanese words end in vowels or /n/ |
| **Morphology** | How words are built from smaller units | "un-believe-able" = 3 morphemes |
| **Agglutinative** | Languages that stack many affixes | Turkish, Huastecan |
| **Isolating** | Languages with few affixes | Chinese, Vietnamese |
| **Proto-language** | Reconstructed ancestor language | Proto-Mixe-Zoquean (~1500 BCE) |

### Statistical/Bayesian Terms

| Term | Definition | In This Notebook |
|------|------------|------------------|
| **Entropy** | Measure of uncertainty/randomness (bits) | High entropy = many equally-likely options |
| **Initial entropy** | How many different signs can START a segment | Reflects word-initial phonotactics |
| **Final entropy** | How many different signs can END a segment | Reflects word-final constraints |
| **Bigram** | A pair of consecutive signs | MS38-MS20 is a bigram |
| **Unigram** | A single sign and its frequency | MS20 appears 52 times |
| **Prior** | Expected value BEFORE seeing data | "MZ words average 3.3 phonemes" |
| **Likelihood** | Probability of data given hypothesis | P(observed | Mixe-Zoquean) |
| **Posterior** | Updated belief AFTER seeing data | Combines prior + likelihood |
| **Bayes Factor (BF)** | Ratio of evidence for two hypotheses | BF > 10 = strong evidence |
| **Log evidence** | Natural log of marginal likelihood | Higher = better model fit |

### Data Structure Terms

| File | Contents | Structure |
|------|----------|----------|
| `segments.csv` | Sign sequences between MS20 markers | segment_id, tokens, length |
| `unigrams.csv` | Sign frequency counts | sign, count, probability |
| `bigrams.csv` | Sign pair frequencies | sign1, sign2, count |

## 4. The Two Candidate Language Families

### Why Only Two?

Geographic and historical evidence limits viable candidates:

| Family | Geographic Fit | Historical Evidence |
|--------|---------------|--------------------|
| **Proto-Mixe-Zoquean** | ✓ Exact overlap with inscription sites | Olmec loanwords in Mesoamerica |
| **Proto-Huastecan** | ✓ Possible migration route | Vonk (2020) alternative hypothesis |
| ~~Proto-Otomanguean~~ | ✗ Oaxaca highlands, not Gulf Coast | No scholar has proposed this |

### Proto-Mixe-Zoquean (Wichmann 1995)

- **Syllable structure**: (C)V(ʔ/h)(C) — allows glottal stops and /h/ in codas
- **Word-finals**: Constrained — often end in /k/, /h/, /ʔ/
- **Morphology**: Polysynthetic (moderate affixation)
- **Modern descendants**: Mixe, Zoque, Popoluca (Veracruz, Oaxaca, Chiapas)

### Proto-Huastecan (Kaufman & Norman 1984)

- **Syllable structure**: CVC — consonant-vowel-consonant
- **Word-finals**: Highly constrained — consonant-final words
- **Morphology**: Agglutinative (many affixes, ergative-absolutive)
- **Modern descendants**: Huastec, Chicomuceltec (northern Veracruz, San Luis Potosí)

## 6. Important Caveats

⚠️ **This analysis is EXPLORATORY, not conclusive.**

1. **Priors are hypothetical** — derived from typological generalizations, not actual corpus statistics from reconstructed lexicons
2. **Corpus limitations** — qW ≈ 0.23 >> 0.1 means even distributional methods face constraints
3. **Script ≠ Language** — sign patterns may reflect writing conventions, not just phonotactics
4. **MS20 assumption** — treating MS20 as a word boundary is a hypothesis, not proven

````python
# =============================================================================
# PDF PARSING FUNCTIONS
# =============================================================================

def extract_text_from_pdf(pdf_path: str) -> str:
    """Extract all text from a PDF file."""
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text


def extract_tables_from_pdf(pdf_path: str) -> List[pd.DataFrame]:
    """Extract tables from a PDF file."""
    tables = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_tables = page.extract_tables()
            for table in page_tables:
                if table and len(table) > 1:
                    df = pd.DataFrame(table[1:], columns=table[0])
                    tables.append(df)
    return tables


def parse_sign_tokens_from_text(text: str) -> List[str]:
    """
    Extract MS-code sign tokens from text.

    Looks for patterns like MS1, MS20, MS101, etc.
    """
    # Pattern: MS followed by 1-3 digits
    pattern = r'MS\d{1,3}'
    tokens = re.findall(pattern, text)
    return tokens


def parse_coordinate_data(text: str) -> pd.DataFrame:
    """
    Parse coordinate-based sign data from article text.

    Expected format in articles:
    LM.1.1: MS38 MS101 MS63 MS20
    LM.1.2: MS149 MS42 ...
    """
    rows = []

    # Pattern: TEXT.COL.ROW: SIGNS or TEXT.COL: SIGNS
    coord_pattern = r'([A-Z]{2})\.([\d]+)(?:\.([\d]+))?[:\s]+([MS\d\s]+)'

    for match in re.finditer(coord_pattern, text):
        text_id = match.group(1)
        col = match.group(2)
        row = match.group(3) if match.group(3) else "1"
        signs = match.group(4).strip()

        # Extract individual MS codes
        sign_tokens = parse_sign_tokens_from_text(signs)

        if sign_tokens:
            rows.append({
                'text_id': text_id,
                'column': int(col),
                'row': int(row),
                'signs': ' '.join(sign_tokens)
            })

    return pd.DataFrame(rows)


print("PDF parsing functions defined.")
````

````python
# =============================================================================
# PARSING FUNCTIONS
# =============================================================================

def parse_sign_sequences(text_data: str, text_id: str) -> List[Dict]:
    """
    Parse sign sequence text into structured segments.
    Segments are delimited by MS20 (per Macri R52).
    """
    segments = []
    segment_id = 0

    for line in text_data.strip().split('\n'):
        line = line.strip()
        if not line:
            continue

        tokens = line.split()

        # Remove trailing MS20 for segment content
        if tokens and tokens[-1] == 'MS20':
            segment_tokens = tokens[:-1]
        else:
            segment_tokens = tokens

        if segment_tokens:
            segments.append({
                'segment_id': f"{text_id}_{segment_id:03d}",
                'text_id': text_id,
                'ms_tokens': ' '.join(segment_tokens),
                'segment_len_tokens': len(segment_tokens),
                'has_ms20_ending': 'MS20' in tokens
            })
            segment_id += 1

    return segments


def extract_all_tokens(text_data: str) -> List[str]:
    """Extract all tokens (signs) from text data."""
    tokens = []
    for line in text_data.strip().split('\n'):
        tokens.extend(line.split())
    return tokens


print("Parsing functions defined.")
````

````python
# =============================================================================
# CREATE CSV FILE 1: isthmus_segments_by_MS20.csv
# =============================================================================

if not DATA_LOADED:
    print("Creating isthmus_segments_by_MS20.csv...")
    print("=" * 60)

    all_segments = []

    # Process each text
    texts = [
        (LA_MOJARRA_SIGNS, "LM", "La Mojarra Stela 1"),
        (LA_MOJARRA_EXTENDED, "LM_EXT", "La Mojarra Extended"),
        (TUXTLA_SIGNS, "TS", "Tuxtla Statuette"),
        (FELDSPAR_SIGNS, "FM", "Feldspar Mask"),
        (CERAMIC_MASK_SIGNS, "CM", "Ceramic Mask"),
        (CHIAPA_DE_CORZO_SIGNS, "CD", "Chiapa de Corzo"),
    ]

    for text_data, text_id, text_name in texts:
        segments = parse_sign_sequences(text_data, text_id)
        all_segments.extend(segments)
        print(f"  {text_name}: {len(segments)} segments")

    # Create DataFrame
    segments_df = pd.DataFrame(all_segments)

    # Add non_eroded_count column
    segments_df['non_eroded_count'] = segments_df['ms_tokens'].apply(
        lambda x: sum(1 for t in str(x).split() if not t.startswith('ERODED'))
    )

    print(f"\nTotal segments: {len(segments_df)}")
    print(f"Mean segment length: {segments_df['segment_len_tokens'].mean():.2f} signs")
    print(f"Max segment length: {segments_df['segment_len_tokens'].max()} signs")

    # Save to CSV
    segments_df.to_csv('isthmus_segments_by_MS20.csv', index=False)
    print("\n✓ Saved: isthmus_segments_by_MS20.csv")
````

````python
# =============================================================================
# FEATURE EXTRACTION FUNCTIONS
# =============================================================================

def extract_segment_features(segments_df):
    """Extract statistical features from segment data."""
    seg_lens = segments_df['segment_len_tokens'].values
    return {
        'seg_len_mean': np.mean(seg_lens),
        'seg_len_std': np.std(seg_lens),
        'seg_len_cv': np.std(seg_lens) / np.mean(seg_lens),
        'seg_len_max': np.max(seg_lens),
        'max_seg_ratio': np.max(seg_lens) / np.mean(seg_lens),
        'seg_len_skew': stats.skew(seg_lens),
    }


def extract_positional_features(segments_df):
    """Extract word-initial and word-final position features."""
    initial_signs, final_signs = [], []

    for _, row in segments_df.iterrows():
        tokens = [t for t in str(row['ms_tokens']).split() if not t.startswith('ERODED')]
        if tokens:
            initial_signs.append(tokens[0])
            final_signs.append(tokens[-1])

    def entropy(items):
        if not items: return 0.0
        counts = Counter(items)
        probs = [c / len(items) for c in counts.values()]
        return -sum(p * math.log2(p) for p in probs if p > 0)

    return {
        'initial_entropy': entropy(initial_signs),
        'final_entropy': entropy(final_signs),
        'n_unique_initial': len(set(initial_signs)),
        'n_unique_final': len(set(final_signs)),
    }


def extract_bigram_features(bigrams_df):
    """Extract bigram distribution features."""
    counts = bigrams_df['count'].values
    total = np.sum(counts)
    probs = counts / total
    sorted_counts = np.sort(counts)[::-1]

    return {
        'bigram_entropy': -np.sum(probs * np.log2(probs + 1e-10)),
        'bigram_types': len(counts),
        'top_10_concentration': np.sum(sorted_counts[:10]) / total,
    }


def extract_frequency_features(unigrams_df):
    """Extract sign frequency distribution features."""
    counts = unigrams_df['count'].values
    total = np.sum(counts)
    sorted_counts = np.sort(counts)[::-1]

    return {
        'n_sign_types': len(counts),
        'freq_concentration': np.sum(sorted_counts[:10]) / total,
        'hapax_ratio': np.sum(counts == 1) / len(counts),
    }


print("Feature extraction functions defined.")
````

````python
# =============================================================================
# BAYESIAN EVIDENCE COMPUTATION
# =============================================================================

def log_evidence_normal(observed, prior_mean, prior_std, lik_std):
    """Log evidence for Normal prior comparison."""
    total_var = prior_std**2 + lik_std**2
    log_p = -0.5 * math.log(2 * math.pi * total_var)
    log_p -= 0.5 * (observed - prior_mean)**2 / total_var
    return log_p


FEATURE_UNCERTAINTIES = {
    'seg_len_mean': (0.5, 0.3),
    'seg_len_cv': (0.15, 0.1),
    'max_seg_ratio': (0.5, 0.3),
    'initial_entropy': (0.5, 0.3),
    'final_entropy': (0.5, 0.3),
    'bigram_entropy': (0.8, 0.5),
    'freq_concentration': (0.1, 0.08),
}

# Scaling factor
scaling_factor = observed['seg_len_mean'] / 3.45
print(f"Scaling factor: {scaling_factor:.3f}")
````

````python
# =============================================================================
# COMPUTE EVIDENCE FOR EACH FAMILY
# =============================================================================

print("=" * 70)
print("BAYESIAN EVIDENCE COMPUTATION")
print("=" * 70)

results = {}

for family, priors in PHONOTACTIC_PRIORS.items():
    print(f"\n{family}:")
    print("-" * 50)

    scaled_priors = priors.copy()
    scaled_priors['seg_len_mean'] = priors['seg_len_mean'] * scaling_factor

    feature_evidence = {}
    total_log_evidence = 0.0

    for feature in FEATURE_UNCERTAINTIES:
        if feature in scaled_priors and feature in observed:
            prior_mean = scaled_priors[feature]
            prior_std, lik_std = FEATURE_UNCERTAINTIES[feature]
            obs_val = observed[feature]

            log_e = log_evidence_normal(obs_val, prior_mean, prior_std, lik_std)
            weight = FEATURE_WEIGHTS.get(feature, 1.0)
            weighted_log_e = weight * log_e

            feature_evidence[feature] = {
                'prior': prior_mean, 'observed': obs_val,
                'log_e': log_e, 'weighted': weighted_log_e
            }
            total_log_evidence += weighted_log_e

            if feature in ['seg_len_mean', 'final_entropy', 'bigram_entropy']:
                print(f"  {feature:20s}: prior={prior_mean:.2f}, obs={obs_val:.2f}, log_e={log_e:.2f}")

    print(f"  {'TOTAL':20s}: {total_log_evidence:.3f}")
    results[family] = {'total': total_log_evidence, 'features': feature_evidence}
````

````python
# =============================================================================
# MODEL COMPARISON & BAYES FACTORS
# =============================================================================

sorted_results = sorted(results.items(), key=lambda x: x[1]['total'], reverse=True)

print("\n" + "=" * 70)
print("MODEL COMPARISON")
print("=" * 70)

print("\nRankings (higher = better fit):")
for rank, (family, res) in enumerate(sorted_results, 1):
    print(f"  {rank}. {family:25s}: {res['total']:.3f}")

best_family, best_res = sorted_results[0]

print("\n" + "=" * 70)
print("BAYES FACTORS")
print("=" * 70)
print("\nInterpretation: BF > 100 = Decisive, > 10 = Strong, > 3 = Moderate")
print(f"\nRelative to {best_family}:")

for family, res in sorted_results[1:]:
    log_bf = best_res['total'] - res['total']
    bf = math.exp(min(log_bf, 700))
    strength = "DECISIVE" if bf > 100 else "STRONG" if bf > 10 else "MODERATE" if bf > 3 else "WEAK"
    print(f"  vs {family:25s}: BF = {bf:,.2f} ({strength})")
````

# PART 6: CRITICAL ASSUMPTIONS
---

## The Fundamental Problem: Script Features ≠ Language Features

This analysis measures properties of the **Isthmian script**, then assumes they reflect properties of the **underlying language**. This requires a chain of unproven assumptions.

### The Assumption Chain

```
SCRIPT OBSERVATION          LINGUISTIC ASSUMPTION           VALIDITY
─────────────────────────────────────────────────────────────────────────
MS20 appears frequently  →  MS20 marks word boundaries   →  UNCERTAIN
                            (could be phrase/clause/verse)

Signs before MS20        →  Represent word-final sounds  →  UNCERTAIN
                            (could be logograms = whole words)

Low final entropy        →  Constrained word-finals      →  UNCERTAIN
                            (could be scribal convention)

Segment length           →  Word length in phonemes      →  UNCERTAIN
                            (segments might be phrases)
```

## What Does MS20 Actually Mark?

### Evidence from R52 (Macri 2017)

Macri's analysis suggests MS20 is a **boundary marker**, but the type of boundary is uncertain:

> "MS20 functions as a boundary or punctuation mark and **may have no syllabic or logographic reading**"

> "Its absence from the second sequence occurs presumably without a change in meaning, suggesting that here it **marks a graphic boundary**"

### Possible Interpretations of MS20

| Interpretation | Linguistic Implication | Evidence |
|----------------|----------------------|----------|
| **Word boundary** | Segments = words | Assumed by this model |
| **Phrase boundary** | Segments = phrases | Equally plausible |
| **Clause boundary** | Segments = clauses | Possible |
| **Verse/poetic marker** | No direct linguistic meaning | Possible |
| **Scribal convention** | Graphic, not linguistic | R52 suggests this |

### Key Problem: MS20 is Absent from Feldspar Mask

R52 states: "MS20 is completely absent from the Feldspar Mask, a text of over 90 signs."

If MS20 marked linguistic word boundaries, its absence would imply:
- The Feldspar Mask has no word boundaries (impossible)
- OR MS20 is a **scribal/textual convention**, not a linguistic marker

This strongly suggests MS20 is a **graphic convention** that varies between texts/traditions.

## How Each Feature Depends on MS20 Assumptions

| Feature | What We Compute | What We Assume | If MS20 ≠ Word Boundary |
|---------|-----------------|----------------|-------------------------|
| `seg_len_mean` | Mean tokens between MS20 | = Mean word length | Measures phrase length instead |
| `seg_len_cv` | Variation in segment length | = Word length variation | Meaningless |
| `initial_entropy` | Entropy of segment-initial signs | = Word-initial phonotactics | Measures phrase-initial patterns |
| `final_entropy` | Entropy of signs before MS20 | = Word-final phonotactics | **Completely invalid** |
| `bigram_entropy` | Sign transition predictability | = Phonotactic constraints | Still partially valid |
| `freq_concentration` | Top-10 sign frequency | = Common syllables | Still valid |

### Most Affected: `final_entropy`

The `final_entropy` feature is the **most problematic** because:

1. It directly depends on MS20 marking word boundaries
2. Our priors assume:
   - Proto-MZ: `final_entropy = 2.0` (constrained by -k, -h, -ʔ codas)
   - Proto-Huastecan: `final_entropy = 1.8` (very constrained CVC)
3. But if MS20 marks phrases, this comparison is meaningless

## The Mixed Script Problem

R53 states: "the script is a mixture of phonetic and logographic signs"

This creates additional complications:

```
SYLLABIC SIGN:     MS38 = /ta/ (hypothetical)
                   → Contributes to phonotactic statistics

LOGOGRAPHIC SIGN:  MS101 = "king" (hypothetical)  
                   → Does NOT reflect phonotactics
                   → Distorts frequency/entropy calculations
```

### Implication for Priors

Our phonotactic priors assume:
- Proto-MZ has (C)V(ʔ/h)(C) syllable structure
- Proto-Huastecan has CVC syllable structure

But if 40% of signs are logograms (R53 estimates ~75% of sign types are rare, suggesting many logograms), then:
- Segment length ≠ phoneme count
- Frequency distributions reflect topic, not phonotactics
- The comparison becomes **topic analysis**, not **phonotactic analysis**

## Validation Proposal: Test on Known Scripts First

Before trusting results on Isthmian, the methodology should be **validated** on scripts where:
1. The language is KNOWN
2. Word boundaries are UNDERSTOOD
3. Sign values are DECIPHERED

# APPENDIX B: Maya Script Comparison — Evidence for MS20 Interpretation

## The Critical Question: What Does MS20 Mark?

The validity of our Bayesian analysis depends entirely on what MS20 represents. The assumption chain is:

```
ASSUMPTION CHAIN (each link is uncertain):

1. MS20 marks word boundaries
   └── Could be phrase boundaries, clause boundaries, or verse markers
   
2. The sign BEFORE MS20 represents the final SOUND of a word
   └── Could be a logogram (whole word), not a syllable
   
3. Therefore, "final_entropy" reflects word-final phonotactics
   └── Could just reflect scribal conventions
   
4. Therefore, low final_entropy = constrained word-finals = CVC language
   └── This is the weakest link!
```

To evaluate this assumption, we compare Isthmian to the **Maya script** — the only fully deciphered Mesoamerican writing system.

## Key Finding: Maya Script Has NO Word Boundary Marker

From scholarly sources:

> "The use of punctuation marks **has not been documented** for classic Maya hieroglyphic writing so far."
> — *ClassicMayan Portal, Research Note 19*

> "In modern transcriptions, punctuation aids clarity... **compensating for the script's lack of explicit punctuation marks.**"
> — *Wikipedia, Maya Script*

### Comparison Table: Maya vs. Isthmian

| Feature | Maya Script | Isthmian Script |
|---------|-------------|----------------|
| Word boundary marker | **NONE** | MS20 (disputed function) |
| Visual word unit | "Glyph block" | Column? |
| Word structure | CVC (consonant-final) | Unknown |
| Script type | Logosyllabic (deciphered) | Logosyllabic (assumed) |
| Syllable signs | CV with "echo vowel" | Unknown |
| Total signs | ~800 (300-500 in use) | ~175 |

## How Maya Handles CVC Words Without Boundary Markers

Maya languages have **CVC structure** (consonant-vowel-consonant), but the script uses **CV syllables**:

```
Maya word: [kah] "fish fin" (CVC)

Written as: ka-ha (CV-CV)
            ↑  ↑
           ka  ha (final vowel is SILENT)

The "echo vowel" repeats the previous vowel to represent the final consonant.
```

### The GLYPH BLOCK Itself Marks Word Boundaries

```
Maya Glyph Block:
┌─────────────┐
│  [main]     │  ← One block = one word/morpheme
│   [affix]   │     (visual grouping, no marker needed)
└─────────────┘

Isthmian (hypothesized):
MS38 MS101 MS44 [MS20] MS63 MS22 [MS20]
                  ↑                  ↑
            boundary?          boundary?
```

### Maya Phonotactics in Script

From Wikipedia:

> "Most Mayan words **end with consonants**, and there may be sequences of two consonants within a word... When these final consonants were sonorants (l, m, n) or gutturals (j, h, ʼ) they were sometimes ignored ('underspelled'). More often, final consonants were written, which meant that an extra vowel was written as well."

This shows that **even in a deciphered script**, the relationship between script patterns and phonotactics is complex.

## Implications for MS20

Since Maya (a related Mesoamerican script) has **no boundary marker**, MS20 in Isthmian is:

1. **NOT inherited from a shared Maya-Isthmian tradition**
2. **Either a unique Isthmian innovation OR serves a different purpose**

### Three Competing Interpretations of MS20

| Interpretation | Proponent | MS20 Function | Implication for Analysis |
|----------------|-----------|---------------|-------------------------|
| **Grammatical suffix** | Justeson & Kaufman (1993) | *-wu* (completive/relativizer) | Final entropy = "what roots take -wu" (**morphology**, not phonotactics) |
| **Phrase/clause boundary** | Macri (partial) | End-of-phrase marker | Final entropy = "what words end phrases" (**syntax**, not phonotactics) |
| **Graphic marker** | Macri (partial) | Column/verse divider | Final entropy = "scribal convention" (**NO linguistic relevance**) |

### Evidence from Macri (2017) R52

> "Its absence from the second sequence occurs presumably **without a change in meaning**, suggesting that here it **marks a graphic boundary**"

This suggests MS20 is **optional** in some contexts — more consistent with a graphic marker than a grammatical suffix.

> "MS20 is completely **absent from the Feldspar Mask**, a text of over 90 signs"

If MS20 were a common grammatical suffix, we'd expect it in ALL texts. Its complete absence suggests either:
- Different scribal tradition (graphic marker hypothesis)
- Different text genre (clause boundary hypothesis)
- Different grammatical constructions (suffix hypothesis)

## Segment Length Analysis: Testing the Word Hypothesis

If MS20 marks **word boundaries**, segment length should match word length predictions.

### Observed Segment Statistics

| Source | Mean Segment Length | Interpretation |
|--------|---------------------|----------------|
| Our data (incomplete) | **~5 signs** | Consistent with WORDS |
| R52 calculation (700 tokens / 52 MS20) | **~12.5 tokens** | Would suggest PHRASES |

### Comparison to Maya Word Length

Maya CVC word written in logosyllabic script:

```
Word: [balam] "jaguar" (CVCVC)

Logographic:  BALAM     (1 sign)
       OR
Syllabic:     ba-la-ma  (3 signs, final vowel silent)
       OR
Mixed:        BALAM-ma  (2 signs, with phonetic complement)
```

**Typical Maya glyph block = 1-4 signs for one word**

### What Proto-MZ Would Predict

Proto-Mixe-Zoquean syllable structure: **(C)V(ʔ/h)(C)**

```
Example Proto-MZ words (from Wichmann 1995):
  *po:m "copal" = CV:C = 1-2 signs if syllabic
  *kah  "fish"  = CVC  = 1-2 signs if syllabic
  *tzaʔ "salt"  = CVʔ  = 1-2 signs if syllabic
```

If Isthmian uses CV syllables like Maya, Proto-MZ words would typically be **2-4 signs**.

Our observed mean of **~5 signs per segment** is actually consistent with the **WORD hypothesis**!

## Revised Assumption Framework

### OLD Framework (Problematic)

```

     ↓

     ↓

     ↓

```

## Testable Predictions

### If MS20 marks WORDS:
- Mean segment length should be ~2-4 signs (typical for logosyllabic CV writing of CVC words)
- Initial/final entropy should correlate with phonotactic constraints
- Segment distribution should match word-length distributions in known languages

### If MS20 marks PHRASES:
- Mean segment length should be ~5-15 signs (phrase length)
- Initial signs should include "topic markers" or "sentence starters"
- Final signs might include particles or clitics

### If MS20 marks GRAPHIC UNITS:
- Segment length should correlate with column position
- No correlation with linguistic features expected
- MS20 should appear mainly at column boundaries

### What R52 Tells Us

From Macri (2017):
- 8 of 22 La Mojarra columns end with MS20 (36%)
- MS20 also occurs **within columns** about 3/4 of the time
- This suggests MS20 has **both** graphic AND linguistic functions
